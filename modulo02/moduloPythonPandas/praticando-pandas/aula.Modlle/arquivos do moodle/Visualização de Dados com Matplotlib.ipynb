Visualização de Dados com Matplotlib

Principais Gráficos com Matplotlib:

•
Gráfico de Linha (plt.plot()): ideal para visualizar tendências e padrões ao longo do tempo, como o acompanhamento das vendas diárias ou mensais.

•
Gráfico de Barras (plt.bar(opens in a new tab)()): permite comparar valores entre diferentes categorias, como o total de vendas por tipo de animal ou por vendedor.

•
Histograma (plt.hist()): exibe a distribuição de uma variável numérica, mostrando a frequência de diferentes faixas de valores.

•
Gráfico de Dispersão (plt.scatter()): útil para identificar relações entre duas variáveis numéricas, como o valor da venda em relação à data da compra.

```python
import pandas as pd

import matplotlib.pyplot as plt

import matplotlib.dates as mdates



# Lê o arquivo CSV para um DataFrame

df_petshop = pd.read_csv('/content/petshop_vendas_completo.csv')



# Converte a coluna `Data_Venda` para o tipo datetime

df_petshop['Data_Venda'] = pd.to_datetime(df_petshop['Data_Venda'])



# Calcula o total de vendas por dia

daily_sales = df_petshop.groupby('Data_Venda')['Valor'].sum().reset_index()



# Ordena por 'Data_Venda' em ordem crescente

daily_sales = daily_sales.sort_values('Data_Venda')



# Cria o gráfico de linha

plt.figure(figsize=(12, 6))  # Tamanho maior da figura para melhor legibilidade

plt.plot(daily_sales['Data_Venda'], daily_sales['Valor'], marker='o', linestyle='-')



# Adiciona título e rótulos aos eixos

plt.title('Total de Vendas ao Longo do Tempo')

plt.xlabel('Data da Venda')

plt.ylabel('Valor Total de Vendas')



# Formata os rótulos do eixo x para mostrar apenas mês e ano

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))  # Ajusta o intervalo conforme necessário



# Exibe o gráfico

plt.tight_layout()

plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt

import matplotlib.dates as mdates



# Lê o arquivo CSV para um DataFrame

df_petshop = pd.read_csv('/content/petshop_vendas_completo.csv')



# Agrupar os dados por diferentes categorias e calcular o total de vendas

category_sales = {

    'Animal': df_petshop.groupby('Animal')['Valor'].sum(),

    'Produto': df_petshop.groupby('Produto')['Valor'].sum(),

    'Pagamento': df_petshop.groupby('Pagamento')['Valor'].sum(),

    'Vendedor': df_petshop.groupby('Vendedor')['Valor'].sum()

}



# Criar e exibir gráficos de barras para cada categoria

for category, sales_data in category_sales.items():

    plt.figure(figsize=(10, 6))

    sales_data.plot(kind='bar', rot=0)

    plt.title(f'Total de Vendas por {category}')

    plt.xlabel(category)

    plt.ylabel('Valor Total de Vendas')

    plt.tight_layout()

    plt.show()
```
